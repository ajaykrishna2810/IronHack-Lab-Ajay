### Cleaning and wrangling data

- Some car brand names refer to the same brand. Replace all brand names that contain the word "Dutton" for simply "Dutton". If you find similar examples, clean their names too. Use `loc` with boolean indexing.

- Convert CO2 Emissions from Grams/Mile to Grams/Km

- Create a binary column that solely indicates if the transmission of a car is automatic or manual. Use `pandas.Series.str.startswith` and .

- convert MPG columns to km_per_liter

Note:
<br>Converting Grams/Mile to Grams/Km

1 Mile = 1.60934 Km

Converting Gallons to Liters

1 Gallon = 3.78541 Liters



```python
df
```

```python
## Your Code here

df.Make.str.count("Dutton").sum() # counts how many times (with sum()) the word 'Dutton' occurs in the column 'Make' 

df = df[df['Make'].str.contains('Dutton')] # finding where the word 'Dutton' exists


df['Make'] = df['Make'].replace(['E. P. Dutton, Inc.'], 'Dutton')
df['Make'] = df['Make'].replace(["S and S Coach Company E.p. Dutton"], 'Dutton') 
df['Make'] = df['Make'].replace(["Superior Coaches Div E.p. Dutton"], 'Dutton') # replacing "Superior Coaches Div E.p. Dutton" with "Dutton"

# change the name of a column
df.rename(columns = {'Grams/Mile':'Grams/Km'}, inplace = True) # changed grams per mile to grams per Km

df.loc[:,'CO2 Emission Grams/Mile'] *= 1/1.60934 # multiplied each entry in the coumn 'CO2 Emission Grams/Mile' with 1/1.60934 (run only once)

    
```
